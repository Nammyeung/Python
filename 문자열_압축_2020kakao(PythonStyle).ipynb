```
def compress(text, tok_len):
  words = [text[i:i+tok_len] for i in range(0, len(text), tok_len)]
  print("words : %s" % words)
  res = []
  cur_word = words[0] # 첫번째 단어 선언
  cur_cnt = 1
  print("zip(words, words[1:] + ['']) : %s, %s" % (words, words[1:] + [''])) #words, words[1:] + [''], a를 제외하고 값을 앞으로 민 것
  # i번째와 i+1번째 단어를 비교해 동일한 단어의 갯수를 파악하는 방법
  for a, b in zip(words, words[1:] + ['']):
    print("a : %s, b : %s" % (a, b))
    if a == b:
      cur_cnt += 1 # 단어가 일치하면 단어의 길이를 늘려서 다음을 탐색
    else:
      #res.append(이중 배열을 넣으면 이중 배열이 입력됨, 스택과 다른점)
      res.append([cur_word, cur_cnt]) #a=b가 다를 때까지 list를 탐색한 다음 같은 단어를 res에 삽입.
      print("res에 %s가 담겼습니다." % res[-1][0])
      print("res 현황 : %s" % res)
      cur_word = b # a or b를 담는다.
      cur_cnt = 1 # 다시 길이 원상복구
  return sum(len(word) + (len(str(cnt)) if cnt > 1 else 0) for word, cnt in res) # res(동일한 단어)에서 word와 cnt를 뽑아서)

#solution을 한 줄로 해결
def solution(text):
  print("len(text)/2 : %f" % (len(text)/2))
  print("len(text) : %f" % (len(text)))
  print("list(range(1, int(len(text)/2) + 1)) + [len(text)]) : %s" % (list(range(1, int(len(text)/2) + 1)) + [len(text)]))
  #tok_len이 index역할을 하는건데.....
  return min(compress(text, tok_len) for tok_len in list(range(1, int(len(text)/2) + 1)) + [len(text)])
  #list(seq): Converts a tuple into a list.


a = [
    "aabbaccc",
    "ababcdcdababcdcd",
    "abcabcdede",
    "abcabcabcabcdededededede",
    "xababcdcdababcdcd",
    'aaaaaa',
]

for x in a:
    print("정답 : %s" % solution(x))

#The zip() function returns a zip object, which is an iterator of tuples 
#where the first item in each passed iterator is paired together, 
#and then the second item in each passed iterator are paired together etc.
```
