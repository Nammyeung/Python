```
def check(num):
  
  if(num < 10 and num % 3 == 0):
    print("손뼉 짝")
    return 1
  elif(num >= 10):
    print("int(str(num)[0] : ", int(str(num)[0]))
    print("int(str(num)[1] : ", int(str(num)[1]))
    if(int(str(num)[0]) % 3 == 0 or int(str(num)[1]) % 3 == 0 and int(str(num)[0]) != 0 and int(str(num)[1]) != 0):
      print("손뼉 짝")
      return 1
  return 0

def solution(num):
  answer = 0
  count = num
  for i in range(1, num+1):
    print("손뼉을 친 번호 : ", i)
    if(i % 3 == 0 and i < 10):
      print("손뼉 짝")
      answer = answer + 1
    else:
      answer = answer + check(i)
  print("answer : " , answer)

  return answer

solution(10)
solution(15)
solution(30)
```
