```
def solve(start, goal, lines):
  dx = [-1, 0, 1, 0]
  dy = [0, 1, 0, -1]

  #next_pos가 Queue의 역할을 한다.
  next_pos = []
  next_pos.append(start)
  while next_pos:
    print("next_pos")
    print(next_pos)
    x, y = next_pos.pop(0)
    print("좌표 %d, %d를 방문하였습니다." % (x, y))
    value = lines[x][y]
    print("%d 걸음만큼 걸었습니다." % (value))
    # 4를 넣은 이유는 상,우,하,좌
    for i in range(4):
      temp_x = x + dx[i]
      temp_y = y + dy[i]
      if temp_x < 0 or temp_y < 0 or temp_x >= n or temp_y >= m:
         continue
      
      # 갈수있는 공간을 찾아 temp_x 및 temp_y에 넣는다 : 방향키 작동
      temp_value = lines[temp_x][temp_y]

      #벽은 패스
      if temp_value < 1:
        continue
         
      # 현재 진행 경로가 더 적은 VALUE일 경우
      #if temp_value == 1:
      if temp_value == 1 or value + 1 < temp_value:
          print("걸음을 옮깁니다.")
          lines[temp_x][temp_y] = value + 1
          next_pos.append((temp_x, temp_y))
          print("좌표 %d, %d로 이동하였습니다." % (temp_x, temp_y))
  print("goal[0] : ")
  print(goal[0])
  print("goal[1] : ")
  print(goal[1])
  return lines[goal[0]][goal[1]]

print("미로의 행과 열의 갯수를 차례로 입력해주세요(구분 : 띄어쓰기) :")
n, m = map(int, input().strip().split())
lines = []
for i in range(n):
  print("각 행의 미로 구조를 입력하시오(0 : 통로, 1 : 벽)")
  line = list(map(int, list(input().strip()))) # 리스트의 자료형을 모두 int로 변환해서 저장 가능
  #https://dojang.io/mod/page/view.php?id=2286
  lines.append(line)
  print(lines)

start = (0, 0) #배열 입력을 이런식으로 한다
print("start : ")
print(start)
goal = (n-1, m-1)
print("goal : ")
print(goal)
print(solve(start, goal, lines))  

for i in range(len(lines)):
  print(lines[i])
  print("") 
```
