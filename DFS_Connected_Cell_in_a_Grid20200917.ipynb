```
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the maxRegion function below.
#def maxRegion(grid):

class Node:
  def __init__(self, data):
    self.adjacents = dict()
    self.data = data

class Graph:
  def __init__(self):
    self.nodes = dict()

  def add(self, data):
    self.nodes[data] = Node(data)

  def connect(self, n1, n2):
    self.nodes[n1].adjacents[n2] = self.nodes[n2]
    self.nodes[n2].adjacents[n1] = self.nodes[n1]

def getBiggestRegion(grid, m):
    check = [[0 for col in range(m)] for row in range(len(grid))]
    print("그래프 객체 g 생성")
    g = Graph() 
    for i, v in enumerate(grid):
        #i : index, v : 각 행
        print("i, v : ")
        print(i, v)
        for j, w in enumerate(v):
        #j : index, w : 각 행의 값들
            print("j, w : ")
            print(j, w)
            #예 : 첫번째 행 * m(행의 갯수) + 각 행의 값
            #poc : 각 경로에 index를 부여하는데 사용가능
            poc = j+i*m
            check[i][j] = poc
            print("poc = j+i*m 계산")
            print("j,i,m, poc")
            print(j,i,m, poc)

            #해당 지점의 값이 1인경우만(영역으로 포함해야 하는 경우)
            if w == 1:
                g.add(poc)
                u, l, ul, ur = poc-m, poc-1, poc-1-m, poc-m+1
                print("u, l, ul, ur, u//m, l//m, poc//m, l%m, ul//m, ul//m, ur//m")
                print(u, l, ul, ur, u//m, l//m, poc//m, l%m, ul//m, ul//m, ur//m)
                if u >= 0 and grid[u//m][j] == 1:
                    g.connect(poc, u)
                if l >= 0 and l//m == poc//m and grid[l//m][l%m] == 1:
                    g.connect(poc, l)
                if ul >= 0 and ul//m == (poc//m)-1 and grid[ul//m][ul%m] == 1:
                    g.connect(poc, ul)
                if ur >= 0 and ur//m == (poc//m)-1 and grid[ur//m][ur%m] == 1:
                    g.connect(poc, ur)
                    
    max = 0
    visited = dict()
    print(check)

    for i, n in g.nodes.items():
        res = countClustersSize(n, visited)
        max = res if res > max else max
    return max
        
def countClustersSize(n, visited):
    if n.data in visited and visited[n.data]:
        return 0
    visited[n.data] = True
    res = 1
    for i, node in n.adjacents.items():
        res += countClustersSize(node, visited)
    return res


print("The first line contains an integer, n, the number of rows in the matrix, grid.")
n = int(input())

print("The second line contains an integer,m , the number of columns in the matrix.")
m = int(input())

grid = []

for _ in range(n):
    print("Each of the following  lines contains a row of m space-separated integers, grid[i][j].")
    grid.append(list(map(int, input().rstrip().split())))


res = getBiggestRegion(grid,m) 
#res = maxRegion(grid)

print(res)
```
