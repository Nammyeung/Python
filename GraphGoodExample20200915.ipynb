```
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the roadsAndLibraries function below.
def dfs(s,arr,visited):
    visited[s]=True
    cost=0
    print("arr[s] : ")
    print(arr[s])
    for i in arr[s]:
        if(not visited[i]):
            cost+=1
            cost+=dfs(i,arr,visited)
    return cost
# n : library 갯수, m : 연결점 갯수
def roadsAndLibraries(n,m, c_lib, c_road, cities):
    if(c_lib<c_road):
        return c_lib*n
    if(m==0):
        return n*c_lib

    arr=[[0] for i in range(n+1)]
    print("arr :")
    print(arr)
    
    visited=[False for i in range(n+1)]

    print("visited :")
    print(visited)

    for i in range(1,n+1):
        arr[i][0]=i
        print("arr2 : ")
        print(arr)

    for i in range(m):
        print("초기 arr")
        print(arr)
        print("추가 들어갑니다.")
        arr[cities[i][0]].append(cities[i][1])
        print("arr[cities[i][0]].append(cities[i][1])")
        print(arr)
        arr[cities[i][1]].append(cities[i][0])
        print("arr[cities[i][1]].append(cities[i][0])")
        print(arr)
    vis=0
    for i in range(1,n+1):
        if(not visited[i]):
            vis+=dfs(i,arr,visited)
    cost=(((vis)*c_road)+((n-vis)*c_lib))
    return cost

#if __name__ == '__main__':
#    fptr = open(os.environ['OUTPUT_PATH'], 'w')

print("The first line contains a single integer , that denotes the number of queries. :")
q = int(input())

for q_itr in range(q):
    
    print("four space-separated integers that describe the respective values of n, m, Clib and Croad, the number of cities, number of roads, cost of a library and cost of a road.")
    nmC_libC_road = input().split()

    n = int(nmC_libC_road[0])

    m = int(nmC_libC_road[1])

    c_lib = int(nmC_libC_road[2])

    c_road = int(nmC_libC_road[3])

    cities = []

    for _ in range(m):
        #icities에 담는 정보가 무엇인지
        print("Each of the next  lines contains two space-separated integers, u[i] and v[i]")
        cities.append(list(map(int, input().rstrip().split())))
        print("Repair the road between cities list[0] and list[1] at a cost of : ")
        print(cities)

    result = roadsAndLibraries(n,m, c_lib, c_road, cities)
    print(result)

#   fptr.write(str(result) + '\n')

#fptr.close()
```
