```
''' 
Tree 자료구조 구현 부분은 
    https://blog.ilkyu.kr/entry/파이썬에서-Tree-트라이-구현하기
    문서를 참고하였습니다.
'''

class Node(object):
    def __init__(self, length = None):
        self.total_children = {}
        print("self.total_children : %s" % (self.total_children))
        if length != None:
            self.total_children[length] = 1
        self.children = {}


class Tree(object):
    def __init__(self):
        self.head = Node(None)

    #insert의 파라미터에는 self정보와 문자열 s가 들어간다.
    def insert(self, s):
        print("insert function 시작")
        curr_node = self.head
        print("curr_node.total_children : %s" % (curr_node.total_children))
        if len(s) not in curr_node.total_children:
            curr_node.total_children[len(s)] = 1
        else:
            count = curr_node.total_children[len(s)]
            curr_node.total_children[len(s)] = count + 1

        for i in range(0, len(s)):
            suffix_length = len(s) - i - 1
            print("curr_node.children : %s" % (curr_node.children))
            if s[i] not in curr_node.children:
                curr_node.children[s[i]] = Node(suffix_length)
                curr_node = curr_node.children[s[i]]
            else:
                curr_node = curr_node.children[s[i]]
                if suffix_length not in curr_node.total_children:
                    curr_node.total_children[suffix_length] = 1
                else:
                    count = curr_node.total_children[suffix_length]
                    curr_node.total_children[suffix_length] = count + 1

    def get_count_prefix(self, prefix, count_wildcard):
        print("#########################################get_count_prefix함수######################################################")
        curr_node = self.head
        print("prefix : %s" % prefix)
        for c in prefix:
            if c not in curr_node.children:
                return 0
            curr_node = curr_node.children[c]
        print("count_wildcard : %s" % count_wildcard)
        if count_wildcard not in curr_node.total_children:
            return 0
        else:
            return curr_node.total_children[count_wildcard]


def solution(words, queries):
    answer = []

    # Make tree
    forward_tree = Tree()
    reverse_tree = Tree()
    for word in words:
        forward_tree.insert(word)
        reverse_tree.insert(word[::-1])
    #print("forward_tree : %s" % (forward_tree))
    #print("reverse_tree : %s" % (reverse_tree))
    print("word : %s" % (word))
    print("word : %s" % (word[::-1]))

    for query in queries:
        print("query : %s" % (query))
        print("query[-1] : %s" % (query[-1]))
        if query[-1] == '?':
            print("len(%s) : %s, query.find('?') : %s" % (query, len(query), query.find('?')))
            count_wildcard = len(query) - query.find('?')
            print("answer에 추가되었습니다.(query[-1] == ? 인 경우)")
            answer.append(forward_tree.get_count_prefix(query[:-count_wildcard], count_wildcard))
            print("answer = %s" % answer)
        else:
            count_wildcard = query.rfind('?') + 1
            print("answer에 추가되었습니다.(query[-1] == ? 이 아닌 경우)")
            answer.append(reverse_tree.get_count_prefix(query[count_wildcard:][::-1], count_wildcard))
            print("answer = %s" % answer)

    return answer


words = ["frodo", "front", "frost", "frozen", "frame", "kakao"]
queries = ["fro??", "????o", "fr???", "fro???", "pro?"]
print(solution(words, queries))
```

```
class Trie:
    def __init__(self, val, depth=0):
        self.child = {}
        self.num = {}
        self.val = val
        self.depth = depth

    def push(self, word):
        print("##########################push함수 시작(word: %s)##############################" % (word))
        node = self
        while len(word) > 0:
            #get() : Returns the value of the specified key
            print("word : %s" % word)
            node.num[len(word)] = node.num.get(len(word), 0) + 1 #dictionary.get(keyname, value)
            # value :	Optional. A value to return if the specified key does not exist.Default value None
            print("node.num[len(word)] = %s" % node.num[len(word)])
            print("node.num = %s" % node.num)
            if word[0] not in node.child:
                print("node.child : %s, word[0] : %s, word : %s" % (node.child, word[0], word))
                node.child[word[0]] = Trie(word[0], node.depth + 1) # Trie의 val가 word[0]를 담고, 깊이가 1인 node.child를 생성한다.
            node = node.child[word[0]]
            print("node : %s" % node)
            word = word[1:]


    def search(self, word):
        print("##########################search함수 시작(word: %s)##############################" % (word))
        node = self
        while len(word) > 0:
            if word[0] == '?':
                return node.num.get(len(word), 0)
            if word[0] not in node.child:
                return 0
            node = node.child[word[0]]
            word = word[1:]

    def __repr__(self):
        print("##########################repr함수 시작##############################")
#문자를 표시하는 방법
        return 'num: {}; val: {};'.format(str(self.num), self.val)
#example
#age = 36 name = "John" txt = "His name is {1}. {1} is {0} years old."
#print(txt.format(age, name))

def solution(words, queries):
    answer = []
    t1 = Trie('')
    t2 = Trie('')
    for word in words:
        t1.push(word)
        t2.push(word[::-1])
    print("t1 : %s" % t1)
    print("t2 : %s" % t2)

    for query in queries:
        if query[0] != '?':
            val = t1.search(query)
        else:
            val = t2.search(query[::-1])
        answer.append(val)
        print("answer : %s" % answer)
    return answer

words = ["frodo", "front", "frost", "frozen", "frame", "kakao"]
queries = ["fro??", "????o", "fr???", "fro???", "pro?"]
print(solution(words, queries))   
```
